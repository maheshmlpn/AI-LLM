```python
import feedparser
import requests

for rss_feed in rss_feeds:
  print(f"Will try to parse rss feed : {rss_feed}")
  rss1 = requests.get(rss_feed).content.decode("utf-8")
  rss1 = rss1.replace("utf-8","unicode")
  p = feedparser.parse(rss1)
  entries = p['entries']
  print(entries)

  for entry in entries:
    print(f"Title : {entry.title}, Link : {entry.link}")
```
